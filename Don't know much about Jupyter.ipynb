# Why is this talk about ipython notebook? Isn't it Jupyter Notebook now?

Jupyter is the evolution of the ipython notebook, and is both an expansion of the original purpose and a commercial support company (the product remains open source). Jupyter is "not really an acronym" but is meant to echo its' expanded language support for
* Julia
* Python
* R

Jupyter also makes more of an effort to separate the web service from the programming language kernels.

## Yeah, the question stands, why is this about ipython?

Because 
* Package managers for Red Hat and Debian families use ipython(3)-notebook.
* The command line to start the notebook is <code>ipython[3] notebook</code>.
* NumPy and SciPy have included ipython for years.
* I only use it for python so far.
* The new shiny needs some tarnish to please my old ashtabula.
